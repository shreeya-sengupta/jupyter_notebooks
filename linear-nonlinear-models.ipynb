## H1: Linear Models - linear hypothesis is a straight line

```python
print("Linear Models - graphical representation")
import matplotlib.pyplot as plt
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Linear hypothesis - straight line')
plt.show()
```

```python
print("Linear hypothesis - curved lines")
from math import pow
import ipywidgets as widgets
from ipywidgets import Layout
style = {'description_width': 'initial'}
n = widgets.Text(value='',description='Enter 5 powers of x separated by comma',
                 style=style,
                 disabled=False,layout=Layout(width='50%', height='80px', positioning='right'))
display(n)
```

## H2: Linear Models - linear hypothesis fits curvatures also

```python
n1 = n.value.split(",")
print('Powers = ', n1)

for j in range(len(n1)):
    print('Current power = ', n1[j])
    print("x = ", x)
    for i in range(len(x)):
        tempX = x[i]
        tempXPow = pow(float(tempX), float(n1[j]))
        y[i] = tempXPow
    print("y = ", y)
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('h(x)')
    plt.title('Linear hypothesis - curvatures')
    plt.show()
```
