```python
print("Linear Models - graphical representation")
import matplotlib.pyplot as plt
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Linear hypothesis - straight line')
plt.show()
```

```python
print("Linear hypothesis - curved lines")
from math import pow
n = input("Input power of x: ")
print("Power of x =", n)
```

```python
print("x = ", x)
for i in range(len(x)):
    tempX = x[i]
    tempXPow = pow(float(tempX), float(n))
    y[i] = tempXPow
print("y = ", y)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Linear hypothesis - curvatures')
plt.show()
```
